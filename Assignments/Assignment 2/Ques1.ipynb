```python
def readFile(name):
    file = open(name)
    text = file.read()
    return text
def characterTokenize(text, k):
    bicharacters = set()
    for i in range(len(text)-k+1):
        bicharacters.add(text[i:i+k])
    return bicharacters

def wordTokenize(text, k):
    words = text.split(" ")
    wordSplit = [words[i:i+k] for i in range(len(words)-k+1)]
    return set(tuple(row) for row in wordSplit)
```
